```python
for dayint in range(1, 2):
    day = '{:02d}'.format(dayint)
    month = "10"
    timestamp = "2020-"+month+"-"+day+"T00:00:00Z"
    print(timestamp)
    with open('csse_covid_19_data/csse_covid_19_daily_reports/'+month+'-'+day+'-2020.csv', 'r') as dailyFile:
        reader = csv.DictReader(dailyFile, delimiter=',')
        i = 0
        event = {}
        for row in reader:
            if row['Country_Region'] not in event:
                event[row['Country_Region']] = {}
            if row['Province_State'] not in event[row['Country_Region']]:
                if row['Province_State'] != 'Recovered':
                    event[row['Country_Region']][row['Province_State']] = {
                        'Country': row['Country_Region'],
                        'Province': row['Province_State'],
                        'Location': row['Country_Region'] + " - " + row['Province_State'],
                        'Confirmed': int(row['Confirmed']),
                        'Deaths': int(row['Deaths']),
                        'Recovered': int(row['Recovered'])
                    }
            else:
                event[row['Country_Region']][row['Province_State']]['Confirmed'] = event[row['Country_Region']][row['Province_State']]['Confirmed'] + int(row['Confirmed']) 
                event[row['Country_Region']][row['Province_State']]['Deaths'] = event[row['Country_Region']][row['Province_State']]['Deaths'] + int(row['Deaths']) 
                event[row['Country_Region']][row['Province_State']]['Recovered'] = event[row['Country_Region']][row['Province_State']]['Recovered'] + int(row['Recovered']) 

            if row['Province_State'] != "":
                if "" not in event[row['Country_Region']]:
                    event[row['Country_Region']][""] = {
                        'Country': row['Country_Region'],
                        'Province': "",
                        'Location': row['Country_Region'] + " - ",
                        'Confirmed': int(row['Confirmed']),
                        'Deaths': int(row['Deaths']),
                        'Recovered': int(row['Recovered'])
                    }
                else:
                    event[row['Country_Region']][""]['Confirmed'] = event[row['Country_Region']][""]['Confirmed'] + int(row['Confirmed']) 
                    event[row['Country_Region']][""]['Deaths'] = event[row['Country_Region']][""]['Deaths'] + int(row['Deaths']) 
                    event[row['Country_Region']][""]['Recovered'] = event[row['Country_Region']][""]['Recovered'] + int(row['Recovered']) 


        fevents = []

        for country in event.keys():
            for province in event[country].keys():
                fevent = event[country][province].copy()
                fevent['timestamp'] = timestamp
                fevents.append(fevent)

        print(len(fevents))
        j = 1
        for fevent in fevents:
            #print(json.dumps(fevent))
            print(str(j)+",", end =" ")
            sbs.send_event(EVENTHUB_NAME, json.dumps(fevent))
            time.sleep(0.1)
            j += 1
```
